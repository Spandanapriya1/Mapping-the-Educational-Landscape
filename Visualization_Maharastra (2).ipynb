```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\SPANDANA PRIYA\Downloads\Maharastra_csv_file.csv")
```

```python
data['College_Fees'] = data['College_Fees'].str.replace(",", "").astype(float)
```

Average College Ratings by Ownership Type

```python
plt.figure(figsize=(10, 6))
avg_ratings = data.groupby('College_Ownership')['College_Rating'].mean().sort_values()

sns.barplot(x=avg_ratings.values,y=avg_ratings.index,hue=avg_ratings.index,palette="viridis",dodge=False,legend=False)

plt.title("Average College Ratings by Ownership Type", fontsize=16)
plt.xlabel("Average Rating", fontsize=12)
plt.ylabel("Ownership Type", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='College_Fees', y='College_Rating', hue='College_Ownership', data=data, palette="cool")
plt.title("Fees vs Ratings", fontsize=16)
plt.xlabel("College Fees (INR)", fontsize=12)
plt.ylabel("College Ratings", fontsize=12)
plt.legend(title="Ownership")
plt.show()
```

```python
ownership_counts = data['College_Ownership'].value_counts()
plt.figure(figsize=(8, 8))
ownership_counts.plot.pie(autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Distribution of Colleges by Ownership Type", fontsize=16)
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['College_Rating'], kde=True, bins=10, color="teal")
plt.title("Distribution of College Ratings", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(x='College_Ownership',y='College_Fees',data=data,hue='College_Ownership',palette="Set2",dodge=False,legend=False)

plt.title("College Fees by Ownership Type", fontsize=16)
plt.xlabel("Ownership Type", fontsize=12)
plt.ylabel("Fees (INR)", fontsize=12)
plt.show()
```

```python
top_rated = data.nlargest(10, 'College_Rating')
plt.figure(figsize=(12, 8))

sns.barplot(x='College_Rating',y='Name_of_the_College',data=top_rated,hue='Name_of_the_College', palette="coolwarm",dodge=False, legend=False)

plt.title("Top 10 Colleges by Rating", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("College", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=top_rated, x='College_Rating', y='College_Fees', marker="o", label="Fees")
plt.title("Variation of Fees for Top 10 Rated Colleges", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Fees (INR)", fontsize=12)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
correlation_matrix = data[['College_Rating', 'College_Fees']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix", fontsize=16)
plt.show()
```

```python
top_locations = data['College_Location'].value_counts().head(10).index
top_location_data = data[data['College_Location'].isin(top_locations)]
plt.figure(figsize=(14, 8))
sns.countplot(y='College_Location', hue='College_Ownership', data=top_location_data, palette="Set1")
plt.title("Ownership Distribution for Top 10 Locations", fontsize=16)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Location", fontsize=12)
plt.legend(title="Ownership")
plt.show()
```
